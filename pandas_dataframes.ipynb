### Filtering a DataFrame

```python
ipl.head(2)
```

```python
ipl['City'].unique
```

```python
# find all the final winners
# method -1
mask = ipl['MatchNumber'] == 'Final'        # first we'll find all final matches and we'll get a boolean series
new_df = ipl[mask]                 # stroing all final matches in "new_df"
new_df[['Season','WinningTeam']]              # and from new_df we're fetching season and winning team


#method-2 (one liner)
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
# (ipl['SuperOver'] == 'Y').sum()
# ipl[ipl['SuperOver'] == 'Y'].shape          # shape will return rows and columns i.e., 14,20 -> 14 matches and 20 is the number of fetures(cols) of each row   
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0]) * 100
```

```python
movies.head(2)
```

```python
# movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
# movies['genres'] == 'Action' & movies['imdb_rating'] > 7.5       # error , because movies have more than one genres therefore we need to first split the strings 
# movies['genres'].split('|)       # error because it is in series format so firt convert it in to string -> movies['genres].str


mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)        # do like this or 
# mask1 = movies['genres'].str.contains('Action')                                   # you can use this also
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2]
```
